```python
import textwrap
import re

# define the keyword to search for
sg = "Singapore"
my = 'Malaysia'
us = 'United States'
china = 'China'
eur = 'Europe'

# use regular expressions to find the keyword in the text string
sg = re.findall(sg, text)
my = re.findall(my, text)
us = re.findall(us, text)
china = re.findall(china, text)
eur = re.findall(eur, text)

# print the number of matches found
print("Number of 'Singapore' found:", len(sg))
print("Number of 'Malaysia' found:", len(my))
print("Number of 'United States' found:", len(us))
print("Number of 'China' found:", len(china))
print("Number of 'Europe' found:", len(eur))
```

```python
# define the data
labels = ['Singapore', 'Malaysia', 'US', 'China', 'Europe']
sizes = [len(sg), len(my), len(us), len(china), len(eur)]

# create the bar chart
plt.bar(labels, sizes, width = 0.5, color = 'slategray', alpha = 0.5, align = 'center')

# add labels and title
plt.xlabel('Keywords')
plt.ylabel('Frequency')
plt.title('Keyword Frequency Chart')

for i, freq in enumerate(sizes):
    plt.text(i, freq + 0.5, str(freq), ha='center')

# display the chart
plt.show()
```
